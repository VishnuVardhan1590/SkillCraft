```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("API_SP.POP.TOTL_DS2_EN_csv_v2_3107.csv", skiprows=4)


df = df.loc[:, ~df.columns.str.contains("Unnamed")]
population_2025 = df["2024"].dropna()


plt.figure(figsize=(9, 5))
plt.hist(population_2025, bins=25, color="skyblue", edgecolor="black")


plt.title("Country Population Distribution (2025)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")

plt.tight_layout()
plt.show()
```
